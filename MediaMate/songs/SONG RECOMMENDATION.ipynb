```python
import nltk
nltk.download('punkt')
```

```python
from nltk.stem.porter import PorterStemmer
stemmer=PorterStemmer()
```

```python
def convert(text):
    token=nltk.word_tokenize(text)
    a=[stemmer.stem(w) for w in token]
    return " ".join(a)
```

```python
songs['lyrics'].apply(lambda x:convert(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vector=TfidfVectorizer(max_features=15000,stop_words='english')
```

```python
vectors=vector.fit_transform(songs['lyrics']).toarray()
```

```python
vectors
```

```python
[s for s in vector.get_feature_names_out()]
```

```python
songs
```

```python
similar=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similar[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
index=songs[songs['song']=='Jealousy'].index[0]
```

```python
index
```

```python
distance=similar[index]
```

```python
distance
```

```python
def recommender(sung):
    index=songs[songs['song']==sung].index[0]
    distance=similar[index]
    song_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in song_list:
       print(songs.iloc[i[0]]['song']) 
```

```python
recommender('Jealousy')
```
